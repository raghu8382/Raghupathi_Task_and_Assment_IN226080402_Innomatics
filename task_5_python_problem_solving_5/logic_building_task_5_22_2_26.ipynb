# Problem 1: Social Media – Post Engagement Analyzer
Problem Statement :
Create a function to analyze engagement on a social media post.


```python
def analyze_engagement(likes_list):
    total_likes = 0

    # Loop to calculate total likes
    for like in likes_list:
        total_likes += like

    print("Total Likes:", total_likes)

    # Check if post is viral
    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")


# Example usage
analyze_engagement([200, 300, 150, 400])
```

# Problem 3: Agriculture – Rainfall Adequacy Checker
Problem Statement :
Create a function that analyzes seasonal rainfall data.


```python
def check_rainfall(rainfall_list, required_level):
    total_rainfall = 0
    
    # Loop to calculate total rainfall
    for rain in rainfall_list:
        total_rainfall += rain
    
    # Calculate average
    average_rainfall = total_rainfall / len(rainfall_list)
    
    print("Average Rainfall:", int(average_rainfall))
    
    # Check adequacy
    if average_rainfall >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")


# Example usage
check_rainfall([70, 75, 80, 63], 70)
```

# Problem 4: Social Media – Duplicate Account Detection
Problem Statement :
Create a function to detect duplicate usernames in a system.


```python
def detect_duplicate_accounts(usernames):
    unique_usernames = set(usernames)

    if len(unique_usernames) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# Example usage
detect_duplicate_accounts(["john", "rahul", "john", "sita"])
```

# Problem 6: Agriculture – Premium Crop Price Filter
Problem Statement :
Create a function that filters crop prices above a premium threshold.


```python
def premium_crop_filter(prices):
    premium_crops = []

    # Loop to filter premium crops
    for price in prices:
        if price > 2000:
            premium_crops.append(price)

    print("Premium Crops:", premium_crops)


# Example usage
premium_crop_filter([1500, 2500, 1800, 3200])
```

# Problem 9: E-Learning – Student Attendance Eligibility System
Problem Statement :
Create a function to calculate attendance percentage and check exam eligibility.


```python
def check_attendance(attendance_list):
    total_classes = len(attendance_list)
    attended_classes = 0

    # Loop to count attended classes
    for status in attendance_list:
        if status == "P":   # P = Present
            attended_classes += 1

    # Calculate percentage
    attendance_percentage = (attended_classes / total_classes) * 100

    print("Attendance Percentage:", attendance_percentage)

    # Check eligibility
    if attendance_percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


# Example usage
check_attendance(["P", "P", "A", "P", "P"])
```

# Problem 10: Smart Electricity Bill Analyzer
Problem Statement :
You are building a smart electricity billing system for residential users.
Create a function that calculates the electricity bill and classifies usage level based on units consumed.


```python
def calculate_electricity_bill(units):
    bill = 0

    # Slab 1: First 100 units
    if units <= 100:
        bill = units * 3

    # Slab 2: 101–200 units
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)

    # Slab 3: Above 200 units
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage Classification
    if bill < 500:
        usage_status = "Low Usage"
    elif bill <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"

    return bill, usage_status


# Example usage
total_bill, status = calculate_electricity_bill(250)

print("Total Bill: ₹", total_bill)
print("Usage Status:", status)
```
