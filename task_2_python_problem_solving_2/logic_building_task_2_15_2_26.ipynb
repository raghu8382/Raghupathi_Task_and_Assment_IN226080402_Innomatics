# 1 Unique Words in a Sentence
Problem Statement:
Given a sentence, identify all unique words and display:
1.The count of unique words
2.The unique words themselves


```python
# Step 1: Take input
sentence = input("Enter sentence: ")

# Step 2: Split sentence into words
words = sentence.split()

# Step 3: Convert list to set to remove duplicates
unique_words = set(words)

# Step 4: Print results
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

 # 2 Highest Salary from Employee Data
Problem Statement:
Store employee names and their salaries, and determine which employee has the highest salary.


```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Step 1: Take first employee as highest initially
highest_employee = ""
highest_salary = 0

# Step 2: Loop through dictionary
for name in employees:
    salary = employees[name]
    
    # Step 3: Compare salaries
    if salary > highest_salary:
        highest_salary = salary
        highest_employee = name

# Step 4: Print result
print("Highest Salary:", highest_employee, "-", highest_salary)
```

# 3 Find Maximum and Minimum Values
Problem Statement:
Find the maximum and minimum values from a list of numbers.


```python
numbers = [45, 22, 89, 10, 66]

# Step 1: Initialize max and min with first element
maximum = numbers[0]
minimum = numbers[0]

# Step 2: Loop through the list
for num in numbers:
    
    # Step 3: Check for maximum
    if num > maximum:
        maximum = num
        
    # Step 4: Check for minimum
    if num < minimum:
        minimum = num

# Step 5: Print results
print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

# 4 Count Products Above a Price Threshold
Problem Statement:
Count how many products have a price greater than 1000.


```python
prices = [450, 1200, 899, 1500, 300]

# Step 1: Initialize counter
count = 0

# Step 2: Loop through list
for price in prices:
    
    # Step 3: Check condition
    if price > 1000:
        count = count + 1

# Step 4: Print result
print("Products above 1000:", count)
```

# 5 Calculate Attendance Percentage
Problem Statement:
Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).


```python
attendance = ["P", "P", "A", "P", "P"]

# Step 1: Initialize counter
present_count = 0

# Step 2: Count "P"
for day in attendance:
    if day == "P":
        present_count = present_count + 1

# Step 3: Calculate total days
total_days = 0
for day in attendance:
    total_days = total_days + 1

# Step 4: Calculate percentage
percentage = (present_count / total_days) * 100

# Step 5: Print result
print("Attendance Percentage:", percentage)
```

# 7 Count Character Frequency
Problem Statement:
Count how many times each character appears in a given string.


```python
text = "pythonp"

# Step 1: Create empty dictionary
frequency = {}

# Step 2: Loop through characters
for ch in text:
    
    # Step 3: Check if character already exists
    if ch in frequency:
        frequency[ch] = frequency[ch] + 1
    else:
        frequency[ch] = 1

# Step 4: Print result
print(frequency)
```

# 10 Calculate Average Marks
Problem Statement:
Calculate the average marks of a student from a list of marks.


```python
marks = [70, 75, 80, 65]

# Step 1: Initialize variables
total = 0
count = 0

# Step 2: Loop through marks
for mark in marks:
    total = total + mark
    count = count + 1

# Step 3: Calculate average
average = total / count

# Step 4: Print result
print("Average Marks:", average)
```
