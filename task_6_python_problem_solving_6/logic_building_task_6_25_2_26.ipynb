# Problem Statement 1: Smart Parking Lot Management System
----->Design a function to manage a smart parking lot.
The system should:
1.Accept vehicle entry and exit logs
2.Calculate total parked vehicles
3.Identify peak parking usage
4.Alert if parking exceeds capacity

```python
def smart_parking(capacity, logs):
    current = 0
    peak = 0
    
    for log in logs:
        if log == "IN":
            current += 1
        elif log == "OUT":
            current -= 1
        
        # Track peak usage
        if current > peak:
            peak = current
    
    print("Currently Parked Vehicles:", current)
    print("Peak Parking Usage:", peak)
    
    if current > capacity:
        print("Parking Status: Exceeded Capacity!")
    else:
        print("Parking Status: Available")


# Sample Input
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking(capacity, logs)
```

# Problem Statement 2:Online Food Delivery Time Estimator
----->Create a function that estimates delivery time based on:
1.Distance (km)
2.Weather condition
3.Traffic level
Apply delays dynamically and display final ETA.

```python
def estimate_delivery_time(distance, traffic, weather):
    
    # Base speed (km per hour)
    speed = 30
    
    # Base time calculation
    base_time = (distance / speed) * 60
    
    # Traffic adjustment
    if traffic == "Low":
        base_time += 5
    elif traffic == "Medium":
        base_time += 15
    elif traffic == "High":
        base_time += 25
    
    # Weather adjustment
    if weather == "Cloudy":
        base_time += 5
    elif weather == "Rainy":
        base_time += 14
    
    print("Estimated Delivery Time:", int(base_time), "minutes")


# Sample Input
estimate_delivery_time(8, "High", "Rainy")
```

# Problem Statement 3:Movie Theatre Seat Occupancy Analyzer
----->Build a function that analyzes seat booking data and:
1.Calculates occupancy percentage
2.Determines if show is Housefull
3.Suggests opening additional shows

```python
def seat_occupancy(total_seats, booked_seats):
    
    booked_count = len(booked_seats)
    
    # Calculate percentage
    occupancy = (booked_count / total_seats) * 100
    
    print("Occupancy:", int(occupancy), "%")
    
    # Determine show status
    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
        print("Suggestion: Consider opening additional show")
    else:
        print("Show Status: Seats Available")


# Sample Input
total_seats = 200
booked_seats = [1] * 150

seat_occupancy(total_seats, booked_seats)
```

# Problem Statement 4:Cloud Server Load Classification System
----->Create a function to classify server load based on CPU usage readings.
Rules:
1.Average CPU < 50% → Normal
2.50%–80% → Warning
3.80% → Critical

```python
def classify_server_load(cpu_readings):
    
    total = 0
    
    # Loop to calculate sum
    for value in cpu_readings:
        total += value
    
    # Calculate average
    average = total / len(cpu_readings)
    
    print("Average CPU Load:", int(average), "%")
    
    # Classification
    if average < 50:
        print("Server Status: Normal")
    elif average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Sample Input
readings = [45, 60, 70, 85, 90]
classify_server_load(readings)
```

# Problem Statement 5:Smart Classroom Resource Usage Monitor
----->Design a function that tracks usage of classroom resources (projector, AC,lights) and identifies overuse patterns.

```python
def monitor_resources(resource_usage):
    
    overused = []
    
    # Loop through dictionary
    for resource, hours in resource_usage.items():
        
        if hours > 8:
            overused.append(resource)
    
    # Output results
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_resources(usage)
```

# Problem Statement 6: Online Event Registration Capacity Controller
----->Create a function that manages event registrations by: Tracking registrations

```python
def event_registration_controller(capacity, registrations):
    
    if registrations > capacity:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlisted = 0
        status = "Open"
    
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


# Sample Input
event_registration_controller(100, 105)
```
