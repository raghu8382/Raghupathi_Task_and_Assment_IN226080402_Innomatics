# Problem Statement 1: Employee Performance Bonus Eligibility
Description:
A company evaluates employee performance scores at the end of the year. You are given a dictionary containing employee names and their performance scores.



```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Step 1: Find highest score manually
highest_score = None

for name in employees:
    score = employees[name]
    
    if highest_score is None:
        highest_score = score
    elif score > highest_score:
        highest_score = score

# Step 2: Find employees with highest score
top_performers = []

for name in employees:
    if employees[name] == highest_score:
        top_performers.append(name)

# Step 3: Print output
print("Top Performers Eligible for Bonus:", end=" ")

for i in range(len(top_performers)):
    if i != len(top_performers) - 1:
        print(top_performers[i], end=", ")
    else:
        print(top_performers[i], end=" ")

print("(Score:", highest_score, ")")
```

# Problem Statement 2: Search Query Keyword Analysis
Description:
An e-commerce website stores customer search queries. You are given a search query sentence entered by a user.


```python
query = "Buy mobile phone buy phone online"

# Step 1: Convert to lowercase
query = query.lower()

# Step 2: Remove punctuation
cleaned_query = ""
for ch in query:
    if ch.isalnum() or ch == " ":
        cleaned_query += ch

# Step 3: Split into words
words = cleaned_query.split()

# Step 4: Count frequency
frequency = {}

for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1

# Step 5: Display words searched more than once
result = {}

for word in frequency:
    if frequency[word] > 1:
        result[word] = frequency[word]

print(result)
```

# Problem Statement 3: Sensor Data Validation
Description:
A factory collects sensor readings every hour. Each reading is stored in a list where the index represents the hour and the value represents the sensor reading.


```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for i in range(len(sensor_readings)):
    if sensor_readings[i] % 2 == 0:
        valid_readings.append((i, sensor_readings[i]))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

# Problem Statement 4: Email Domain Usage Analysis
Description:
A company wants to analyze which email providers its users are using. You are given a list of employee email IDs.


```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
domain_count = {}

# Step 1 & 2: Extract domain and count
for email in emails:
    parts = email.split("@")
    domain = parts[1]
    
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

# Step 3: Calculate percentage
total_emails = len(emails)

for domain in domain_count:
    percentage = (domain_count[domain] / total_emails) * 100
    print(domain + ":", int(percentage), "%")
```

# Problem Statement 5: Sales Spike Detection
Description:
A retail company tracks daily sales. Sudden spikes in sales may indicate promotions or unusual activity.


```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Step 1: Calculate average manually
total = 0
for value in sales:
    total += value

average = total / len(sales)

# Step 2: Calculate threshold (30% above average)
threshold = average + (average * 0.3)

# Step 3: Detect spike days
for i in range(len(sales)):
    if sales[i] > threshold:
        print("Day", i + 1, ":", sales[i])
```

# Problem Statement 6: Duplicate User ID Detection
Description:
A system stores user IDs during registration. Duplicate IDs can cause data integrity issues.


```python
# Problem 6: Duplicate User ID Detection

# Input list of user IDs
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Dictionary to store frequency of each user ID
id_count = {}

# Step 1: Count occurrences
for user in user_ids:
    if user in id_count:
        id_count[user] += 1
    else:
        id_count[user] = 1

# Step 2: Display only duplicates
for user in id_count:
    if id_count[user] > 1:
        print(user, "→", id_count[user], "times")
```
