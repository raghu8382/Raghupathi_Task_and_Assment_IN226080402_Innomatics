# 1. User Login Check
Problem Statement:
Given a username and password, check whether login is successful.


```python
# Take username and password input and remove extra spaces
username = input().strip()
password = input().strip()

# Check if entered credentials match predefined values
if username == "admin" and password == "1234":
    print("Login Successful")
else:
    print("Invalid Credentials")
```

# 2. Pass / Fail Analyzer
Problem Statement:
Analyze student results.


```python
# List of student marks
marks = [45, 78, 90, 33, 60]

# Initialize counters
pass_count = 0
fail_count = 0

# Iterate through marks and count pass/fail
for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

# Display results
print("Total Number Of Pass Students :", pass_count)
print("Total Number Of Fail Students :", fail_count)
```

# 3. Simple Data Cleaner
Problem Statement:
Clean and standardize user names.


```python
# List containing names with inconsistent formatting
names = [" Alice ", "bob", " CHARLIE "]

# New list to store cleaned names
new_list = []

# Convert to lowercase and remove extra spaces
for name in names:
    modified = name.lower().strip()
    new_list.append(modified)

# Print cleaned names
print(new_list)
```

# 4. Message Length Analyzer
Problem Statement:
Analyze message sizes.


```python
# List of messages
messages = ["Hi", "Welcome to the platform", "OK"]

# Iterate through each message
for message in messages:
    length = len(message)  # Calculate length of message
    print("Message:", message)
    print("Length of Message:", length)

    # Check if message length is greater than 10
    if length > 10:
        print("This message is longer than 10 characters")

    print()  # Print empty line for better readability
```

# 5. Error Message Detector
Problem Statement:
Detect error messages from system logs.


```python
# List of log messages
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Initialize error counter
error_count = 0

# Count occurrences of "ERROR"
for log in logs:
    if log == "ERROR":
        error_count += 1

# Display total number of errors
print("Number Of Errors :", error_count)
```
