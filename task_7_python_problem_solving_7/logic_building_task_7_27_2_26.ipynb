```python
# ================================================================
# Problem Statement 1: Smart Light Controller
# Create a class to control a smart light that can be turned ON or OFF 
# and display its current status.
# ================================================================
```

```python
class SmartLight:
    
    # Constructor: Initializes light name and default status
    def __init__(self, name):
        self.name = name
        self.status = "OFF"   # Default state
    
    # Method to turn ON the light
    def turn_on(self):
        self.status = "ON"
    
    # Method to turn OFF the light
    def turn_off(self):
        self.status = "OFF"
    
    # Method to display current status
    def display_status(self):
        print(f"{self.name} is {self.status}")


light = SmartLight("Bedroom Light")
light.turn_on()
light.display_status()

```

```python
# ================================================================
# Problem Statement 2: Employee ID Card System
# Create a class that stores employee name, ID, and department
# and displays formatted ID card details.
# ================================================================
```

```python
class EmployeeIDCard:
    
    # Constructor: Stores employee details
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department
    
    # Method to display ID card information
    def display_id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


employee = EmployeeIDCard("Rahul", "EMP102", "IT")
employee.display_id_card()

```

```python
# ================================================================
# Problem Statement 3: Mobile Contact Record
# Create a class to store contact name and phone number
# and display contact information.
# ================================================================
```

```python
class MobileContact:
    
    # Constructor: Initializes contact details
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
    
    # Method to display contact information
    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


contact = MobileContact("Anita", "9876543210")
contact.display_contact()

```

```python
# ================================================================
# Problem Statement 4: Product Price Tag Generator
# Create a class that stores product name and price
# and prints a formatted price tag.
# ================================================================
```

```python
class ProductPriceTag:
    
    # Constructor: Stores product details
    def __init__(self, name, price):
        self.name = name
        self.price = price
    
    # Method to display formatted price tag
    def display_price_tag(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))


product = ProductPriceTag("Headphones", 2499)
product.display_price_tag()

```

```python

# Problem Statement 5: Movie Rating Display System
# Create a class that stores movie name and rating
# and displays movie details with rating out of 5.
# ================================================================
```

```python
class MovieRating:
    
    # Constructor: Stores movie details
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating
    
    # Method to display movie details
    def display_movie_details(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")


movie = MovieRating("Inception", 4.8)
movie.display_movie_details()

```

```python
# ================================================================
# Problem Statement 6: Delivery Address Manager
# Create a class that stores customer name and delivery address
# and prints formatted delivery details.
# ================================================================
```

```python
class DeliveryManager:
    
    # Constructor: Stores delivery details
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address
    
    # Method to display delivery details
    def display_delivery_details(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)


delivery = DeliveryManager("Suman", "Hyderabad")
delivery.display_delivery_details()
```
